# Importando Dados

```python
import tabula
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
lista_tabelas = tabula.read_pdf('Fevereiro_2022.pdf', pages = '11-16')
print('Total de Tabelas: ' + str(len(lista_tabelas))) 
type(lista_tabelas)
```

# Tabela 1

```python
# Importando a tabela da primeira página
df_1 = lista_tabelas[0]
df_1
```

```python
# Removendo colunas desnecessárias
df_1.drop(columns = ['Unnamed: 0', 'Unnamed: 1'], inplace = True)
df_1.columns = ['Dados']

# Incluindo Categoria
df_1['Categoria'] = ''
for i in range(len(df_1)):
    if i in range(5,12):
        df_1['Categoria'].iloc[i] = 'Veículo de Entrada'
    elif i in range(17,29):
        df_1['Categoria'].iloc[i] = 'Hatch Compacto'
    elif  i in range(34,40):
        df_1['Categoria'].iloc[i] = 'Hatch Médio'
    else:
        df_1['Categoria'].iloc[i] = np.nan

# Removendo linhas desnecessárias
df_1.dropna(inplace = True)

# Aplicando pequenas correções
df_1['Dados'] = df_1['Dados'].str.replace('o', '')
df_1['Dados'] = df_1['Dados'].str.replace(' =', '')
```

```python
# Ajustando parametros para divisão dos dados em colunas
df_1['Dados'] = df_1['Dados'].str.replace('YARIS HB', 'YARIS_HB')
df_1['Dados'] = df_1['Dados'].str.replace('HONDA/CITY HATCH', 'HONDA/CITY_HATCH')
df_1['Dados'] = df_1['Dados'].str.replace('CLASSE A', 'CLASSE_A')
df_1['Dados'] = df_1['Dados'].str.replace('CRUZE HB', 'CRUZE_HB')

# Definindo novas colunas
df_1[['Coluna_1', 'Coluna_2', 'Janeiro', 'Fevereiro']] = df_1['Dados'].str.split(' ', expand=True)
df_1[['Fabricante', 'Modelo']] = df_1['Coluna_2'].str.split('/', expand=True)
df_1['Modelo'] = df_1['Modelo'].str.replace('_', ' ')
df_1['Janeiro'] = df_1['Janeiro'].str.replace('.', '').astype(int)
df_1['Fevereiro'] = df_1['Fevereiro'].str.replace('.', '').astype(int)
df_1['Acumulado'] = df_1['Janeiro'] + df_1['Fevereiro']
df_1['Ano'] = 2022

# Removendo e reordenando colunas
df_1.drop(columns = ['Dados', 'Coluna_1', 'Coluna_2'], inplace = True)
df_1 = df_1[['Fabricante', 'Modelo', 'Categoria', 'Janeiro', 'Fevereiro', 'Acumulado', 'Ano']]

df_1.index = range(df_1.shape[0])
df_1
```

# DataFrame Completo

```python
df_final = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6], axis = 0)
df_final.index = range(df_final.shape[0])

# Incluindo o tipo do motor
Lista_Eletricos = ['500E', 'E-JS1', 'ZOE', 'LEAF', 'I3', '330E', '530E', 'MODEL S', 'E TRON GT', 'BOLT']
Lista_Hibridos = ['COROLLA', 'PANAMERA', 'COROLLA CROSS']

df_final['Motor'] = ''
for i in range(len(df_final)):
    if df_final['Modelo'].iloc[i] in Lista_Eletricos:
        df_final['Motor'].iloc[i] = 'Elétrico'
    elif df_final['Modelo'].iloc[i] in Lista_Hibridos:
        df_final['Motor'].iloc[i] = 'Termico ou Hibrido'
    else:
        df_final['Motor'].iloc[i] = 'Termico'

df_final
```
